```python
data = pd.read_csv("/Users/ashtonfremin/Desktop/LURA 2020/SWEA_info.csv")
data_2 = data.drop(columns=['DATE_TIME'])
clean_data = data_2.dropna()
```

```python
# data normalization
clean_data = (clean_data - np.mean(clean_data, axis=0)) / np.std(clean_data, axis=0)
clean_data = clean_data.values
```

```python
# Initialization and training
n_neurons = 140
m_neurons = 100
som = MiniSom(n_neurons, m_neurons, clean_data.shape[1], sigma=1.5, learning_rate=.5, 
              neighborhood_function='gaussian', random_seed=0)

som.pca_weights_init(clean_data)
som.train(clean_data, 1000, verbose=True)  # random training
```

```python
som = MiniSom(140, 100, clean_data.shape[1], sigma=3., learning_rate=.7,
              neighborhood_function='gaussian', random_seed=10)

max_iter = 1000
q_error = []
t_error = []

for i in range(max_iter):
    rand_i = np.random.randint(len(clean_data))
    som.update(clean_data[rand_i], som.winner(clean_data[rand_i]), i, max_iter)
    q_error.append(som.quantization_error(clean_data))
    t_error.append(som.topographic_error(clean_data))

plt.plot(np.arange(max_iter), q_error, label='quantization error')
plt.plot(np.arange(max_iter), t_error, label='topographic error')
plt.ylabel('quantization error')
plt.xlabel('iteration index')
plt.legend()
plt.show()
```
